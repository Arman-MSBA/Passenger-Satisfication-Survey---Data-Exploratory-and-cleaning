# ******* Data Cleansing *******

# Dropping the Class column:

```python
df_Business.drop(['Class'], axis=1, inplace=True)
```

```python
df_Business = df_Business.reset_index(drop=True)
df_Business
```

### We now dropped the 'Class' column since all the customers were business class, so it doesn't really help us derive anything.

# Renaming “df_Business” columns by making all the columns’ names in lowercase.  If the column name has whitespace we replace it with an underscore:

```python
df_Business.rename(columns = lambda x: x.lower().strip().replace(' ','_'), inplace= True)

df_Business.columns
```

### We have changed the column names to all lower space and whenever there is a space, we changed it to an underscore. The strip also removes any space before or after each of the titles.

# Modifying improperly formatted data and handling bad data:

```python
df_Business.info()
```

```python
columns = df_Business.columns
columns

for x in columns:
    print(x)

    unique = df_Business[x].unique()
    print(unique, '\n*************************************\n\n')
```

### We have some columns that have the same values with different words, we will combine them to one. These are in the 'gender', 'customer_type', 'type_of_travel', and 'satisfaction' columns. There are some some extreme outlier values that make no sense.  We change the categorical and out of place numerical values to proper numerical values on a scale of 5 to fit the rest of the values in the dataframe. 

```python
df_Business['gender'] = df_Business['gender'].replace(['F','M'],['Female','Male'])
df_Business['customer_type'] = df_Business['customer_type'].replace(['Loyal','disloyal'],['Loyal Customer','Disloyal Customer'])
df_Business['type_of_travel'] = df_Business['type_of_travel'].replace(['Business'],['Business travel'])
df_Business['satisfaction'] = df_Business['satisfaction'].replace(['Dissatisfied'],['dissatisfied'])
df_Business.at[0,'flight_distance'] = abs(df_Business.at[0,'flight_distance'])
df_Business[df_Business['age'] > 179] = np.NaN
df_Business[df_Business['age'] == 0] = np.NaN
df_Business[df_Business['flight_distance'] > 100000] = np.NaN
df_Business['food_and_drink'] = df_Business['food_and_drink'].replace(['Great'],['4'])
df_Business['inflight_entertainment'] = df_Business['inflight_entertainment'].replace([10],[5])
df_Business['baggage_handling'] = df_Business['baggage_handling'].replace(['A'],['5'])
df_Business['ease_of_online_booking'] = df_Business['ease_of_online_booking'].replace('Unknown', np.NaN)


df_Business
```

```python
for x in columns:
    print(x)

    unique = df_Business[x].unique()
    print('#.of unique values: ', len(unique), '\nList of unique values: ',unique, '\n*************************************\n')
```

# Removing the fully duplicated observations:

```python
df_Business.duplicated().sum()
df_Business.drop_duplicates(inplace=True)
df_Business.duplicated().sum()
df_Business
```

### Here, we dropped all the duplicated rows and re-ran the dataframe to see that they have been dropped.

```python
df_Business.isnull().sum()
```

## At this point the only remaining part of data cleansing we have is fixing the NA values, and here is the table of the number of NA values in each column and how we will convert it:

|  Column                            | N of NA values   | NA values converted to  
|  :------                           |   -----          |   -----:
| gender                             |     3            |  Mode 
| customer_type                      |     3            |  Mode
| age                                |   864            |  Mean
| type_of_travel                     |     3            |  Mode
| flight_distance                    |   833            |  Mean
| inflight_wifi_service              |     3            |  Mode
| departure/arrival_time_convenient  |     3            |  Mode
| ease_of_online_booking             |  5102            |  Mode
| food_and_drink                     |  4675            |  Mode
| inflight_entertainment             |  4370            |  Mode
| baggage_handling                   |  4370            |  Mode
| cleanliness                        |  4370            |  Mode
| departure_delay_in_minutes         |  4370            |  Mean
| arrival_delay_in_minutes           |  5209            |  Mean
| satisfaction                       |  2132            |  Mode

```python
df_Business['gender'].fillna(df_Business['gender'].mode()[0], inplace=True)
df_Business['customer_type'].fillna(df_Business['customer_type'].mode()[0], inplace=True)
df_Business['age'].fillna(int(df_Business['age'].mean()), inplace=True)
df_Business['type_of_travel'].fillna(df_Business['type_of_travel'].mode()[0], inplace=True)
df_Business['flight_distance'].fillna(df_Business['flight_distance'].mean(), inplace=True)
df_Business['inflight_wifi_service'].fillna(df_Business['inflight_wifi_service'].mode()[0], inplace=True)
df_Business['departure/arrival_time_convenient'].fillna(df_Business['departure/arrival_time_convenient'].mode()[0], inplace=True)
df_Business['ease_of_online_booking'].fillna(df_Business['ease_of_online_booking'].mode()[0], inplace=True)
df_Business['food_and_drink'].fillna(df_Business['food_and_drink'].mode()[0], inplace=True)
df_Business['inflight_entertainment'].fillna(df_Business['inflight_entertainment'].mode()[0], inplace=True)
df_Business['baggage_handling'].fillna(df_Business['baggage_handling'].mode()[0], inplace=True)
df_Business['cleanliness'].fillna(df_Business['cleanliness'].mode()[0], inplace=True)
df_Business['departure_delay_in_minutes'].fillna(df_Business['departure_delay_in_minutes'].mean(), inplace=True)
df_Business['arrival_delay_in_minutes'].fillna(df_Business['arrival_delay_in_minutes'].mean(), inplace=True)
df_Business['satisfaction'].fillna(df_Business['satisfaction'].mode()[0], inplace=True)
```

```python
df_Business.isnull().sum()
```

# Fixing the columns datatypes:

```python
df_Business = df_Business.convert_dtypes()
df_Business['ease_of_online_booking'] = pd.to_numeric(df_Business['ease_of_online_booking'])
df_Business['food_and_drink'] = pd.to_numeric(df_Business['food_and_drink'])
df_Business['baggage_handling'] = pd.to_numeric(df_Business['baggage_handling'])

df_Business.info()
```
